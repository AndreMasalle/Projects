# Korelasi 

Data eksplorasi tidaklah cukup dengan mengetahui dimensi data dan statistical properties saja, tetapi kita juga perlu sedikit menggali tentang hubungan atau korelasi dari setiap feature

```python
import pandas as pd

# Read the dataset
dataset = pd.read_csv('https://storage.googleapis.com/dqlab-dataset/pythonTutorial/online_raw.csv')


# Exclude non-numeric columns from correlation calculation
numeric_columns = dataset.select_dtypes(include=['float64', 'int64','bool']).columns
dataset_corr = dataset[numeric_columns].corr()

# Print correlation and other information
print('Korelasi dataset:\n', dataset_corr)
print('Distribusi Label (Revenue):\n', dataset['Revenue'].value_counts())
```

```python
print(dataset_corr.shape)
```

### Korelasi antar 2 variabel

```python
print('Korelasi BounceRates-ExitRates:', dataset_corr.loc['BounceRates', 'ExitRates'])
print('Korelasi Revenue-PageValues:', dataset_corr.loc['Revenue', 'PageValues'])
print('Korelasi TrafficType-Weekend:', dataset_corr.loc['TrafficType', 'Weekend'])
```

### Visualsasi 

```python
import seaborn as sns
import matplotlib.pyplot as plt
# import sklearn.metrics as confusion_matrix, ini tuh buat visualisasi akurasi model wkkwkw
```

```python
f, ax = plt.subplots(figsize = (16,16))
plt.title("Correlation Matrix")
sns.heatmap(dataset_corr, linewidths=0.25, vmax=1.0, square=True, linecolor='black',
             annot=True, cmap=sns.cubehelix_palette())
plt.tight_layout()
plt.show()
```
