```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


table = pd.read_csv("https://storage.googleapis.com/dqlab-dataset/penduduk_gender_head.csv")
table.head()
# x_label = table['NAMA KELURAHAN']
# plt.bar(x=np.arange(len(x_label)),height=table['LAKI-LAKI WNI'])
# plt.show()
```

```python
grouped_data = table.groupby('NAMA KECAMATAN')['LAKI-LAKI WNI'].sum().reset_index()

plt.bar(x=np.arange(len(grouped_data)),height=grouped_data['LAKI-LAKI WNI']) # Buat isi data nya
# np.arange(len(x_label)) # Urutin data nya, dari index paling awal
# height=table['LAKI-LAKI WNI'] # Data di y axis, sekaligus dinamain
plt.xticks(np.arange(len(grouped_data)), grouped_data['NAMA KECAMATAN'], rotation=90) # Namain x-axis
# rotation=90 #Buat rotasi label di x-axis

# Add labels to the axes
plt.xlabel('Kecamatan')
plt.ylabel('Laki-Laki WNI')
plt.title('Persebaran Jumlah Penduduk Laki-laki WNI di Jakarta Pusat')


plt.show()
```
